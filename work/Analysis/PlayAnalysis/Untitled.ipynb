```python
from config import get_engine

from analysis import get_scores

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = get_scores(classification='men 8 end')
df.head()
```

```python
## first end with hammer

first_end_hammer = df.loc[(df.EndNum==1) & (df.Hammer==1)].groupby(['EventID','TeamID'])['EndResult'].mean()
first_end_hammer = first_end_hammer.rename('End1Hammer')
first_end_n_hammer = df.loc[(df.EndNum==1) & (df.Hammer==0)].groupby(['EventID','TeamID'])['EndResult'].mean()
first_end_n_hammer = first_end_n_hammer.rename('End1NoHammer')
```

```python
end1 = pd.concat([first_end_hammer,first_end_n_hammer],axis=1)
```

```python
end1.plot(kind='scatter',x='End1Hammer',y='End1NoHammer')
```

```python
from sklearn.cluster import KMeans
```

```python
end1 = end1.replace(np.nan,0)
```

```python
cls = KMeans(n_clusters=4).fit(end1[['End1Hammer','End1NoHammer']])
```

```python
end1['group'] = cls.predict(end1[['End1Hammer','End1NoHammer']])
```

```python
fig,ax = plt.subplots(1,figsize=(10,10))

groups = []
for group in end1.group.unique():
    temp_df = end1.loc[end1.group==group]
    plt.scatter(temp_df.End1Hammer,temp_df.End1NoHammer)
    groups.append(group)
plt.xlabel('Hammer Expected Score')
plt.ylabel('No Hammer Expected Score')
plt.legend(groups)
plt.show()
```
