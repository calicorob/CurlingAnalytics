```python
## library imports

## data
from analysis import get_win_pcts,get_scores,get_win_pcts_matrix

## visualization
from seaborn import heatmap
import matplotlib.pyplot as plt
```

```python
## helper func
def plot_heatmap(df,**kwargs):
    heatmap(df,**kwargs)
```

```python
## plotting parameters
font_kwargs = {
     'fontsize':14
    ,'fontweight':'bold'
}

names = ['No Hammer','Hammer']
```

## 10 End Women's

```python
## 10 end women's
women_10_ends = [get_win_pcts_matrix(classification='women 10 end',hammer=hammer) for hammer in [0,1]]
```

```python
## no hammer
women_10_ends[0]
```

```python
## hammer
women_10_ends[1]
```

```python
fig,axs = plt.subplots(2,figsize=(10,10))

for i,ax in enumerate(axs.flat):
    plot_heatmap(women_10_ends[i],ax=ax,vmin=0,vmax=1)
    ax.set_xlabel("EndSituation",**font_kwargs)
    ax.set_ylabel("EndNum",**font_kwargs)
    ax.set_title("".join(["Win Percentages\n","Women 10 End\n",names[i]]),**font_kwargs)
    
plt.tight_layout()
plt.show()
```
