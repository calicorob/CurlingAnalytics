```python
## library imports

## db connection params
from config import get_engine

## IO
import pandas as pd

## linear algebra
import numpy as np

## defaultdict
from collections import defaultdict

#pivoting
from pivoting import create_game
```

```python
## db engine
database="Scores"
engine = get_engine(database=database)
```

```python
## scores

scores = pd.read_sql("SELECT * FROM Scores LIMIT 1000;",con=engine)
scores.head()
```

```python
games = []
for game_id in scores.GameID.unique():
    games.append(create_game(game_id=game_id,games=scores))
```

```python
pd.concat([game.to_frame() for game in games if game],axis=0,sort=True)
```
