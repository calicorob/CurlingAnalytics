```python
## library imports

## db connection params
from config import get_engine

## IO
import pandas as pd

## linear algebra
import numpy as np

## defaultdict
from collections import defaultdict

#pivoting
from pivoting import create_game
```

```python
## db engine
database="Scores"
engine = get_engine(database=database)
```

```python
## scores

sql = """

        SELECT
             TeamID AS Team
            ,Hammer
            ,FinalScore
            ,End1
            ,End2
            ,End3
            ,End4
            ,End5
            ,End6
            ,End7
            ,End8
            ,End9
            ,End10
            ,End11
            ,End12
            ,GameID
        FROM Scores
        WHERE GameID IN (
        
            SELECT GameID 
            FROM Scores
            WHERE Team = 'Lawes/Lawes'
        
        );





"""


scores = pd.read_sql(sql,con=engine)
scores.head()
```

```python
scores.loc[scores.GameID == 2229]
```

```python
games = []
for game_id in scores.GameID.unique():
    try:
        games.append(create_game(game_id=game_id,games=scores))
    except:
        pass
```

```python
pd.concat([game.to_frame() for game in games if game],axis=0,sort=True)
```
